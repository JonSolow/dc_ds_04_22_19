### 1. bootstrapping !!
##### Let's say we have a pandas dictionary (our sink) and inside we have one urn (with red and white balls). The urn has 40 red balls 60 white 

##### What's the probability to pick a red ball with replacement 3 times in a row ? 

- Can you answer this question theoretically? 

Pr(Red Ball 3 Times with replacement) = 0.4**3

```python
0.4**3
```

##### lets create a simulation running 1000/10000/100000 iterations and count how many times we got three reds in a row.

`sink : {'urn': {'red': 40, 'white':60}}`

`use np.random.rand`

```python
sink =  {'urn': {'red': 40, 'white': 60}}

balls = np.array(list(sink['urn'].keys()))
probs = np.array(list(sink['urn'].values()))
probs = probs / sum(probs)

def choose_3_balls():
    return np.random.choice(balls, size=3, p=probs)


def simulate_3_balls(size=1000):
    sim_list = []
    for _ in range(size):
        sim_list.append(choose_3_balls())
    return sim_list
        
def check_3_colors(sim_list, check_list = [None, None, None]):
    check_count = 0
    for sims in sim_list:
        if np.all(sims == check_list): check_count += 1
    return check_count
```

```python
def choose_3_balls_alt():
    rand_3 = np.random.random_sample(3)
    selected_balls = ['red' if x<0.4 else 'white' for x in rand_3]
    return selected_balls
```

```python
def simulate_3_balls_alt(size=1000):
    sim_list = []
    for _ in range(size):
        sim_list.append(choose_3_balls_alt())
    return sim_list
```

```python
sims_1000_alt = simulate_3_balls_alt(1_000)
sims_10000_alt = simulate_3_balls_alt(10_000)
sims_100000_alt = simulate_3_balls_alt(100_000)

print(check_3_colors(sims_1000_alt, ['red', 'red', 'red']) / len(sims_1000_alt))
print(check_3_colors(sims_10000_alt, ['red', 'red', 'red']) / len(sims_10000_alt))
print(check_3_colors(sims_100000_alt, ['red', 'red', 'red']) / len(sims_100000_alt))
```

```python
sims_1000 = simulate_3_balls(1_000)
sims_10000 = simulate_3_balls(10_000)
sims_100000 = simulate_3_balls(100_000)

percent_all_red_1000 = check_3_colors(sims_1000, ['red', 'red', 'red']) / len(sims_1000)
percent_all_red_10000 = check_3_colors(sims_10000, ['red', 'red', 'red']) / len(sims_10000)
percent_all_red_100000 = check_3_colors(sims_100000, ['red', 'red', 'red']) / len(sims_100000)


percent_all_red_1000, percent_all_red_10000, percent_all_red_100000
```
