```python
import json
from IPython.display import display, Markdown, Latex
```

```python
from json2table import convert
f = open('brewerydb.json')
data = json.load(f)
build_direction = "LEFT_TO_RIGHT"
table_attributes = {"table-layout": "fixed"} #{"style" : "width:100%"}
html = convert(data, build_direction=build_direction, table_attributes=table_attributes)

display(Markdown(html))
```
