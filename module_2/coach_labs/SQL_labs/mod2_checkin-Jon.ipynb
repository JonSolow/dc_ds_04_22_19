### Import packages 

```python
import sqlite3
import pandas as pd
```

### Connect to database

```python
conn = sqlite3.connect('amazon_db.db')
c = conn.cursor()

c.execute("""SELECT * FROM books; """) 

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df.num_copies = df.num_copies.astype('int')
df_sort = df.sort_values('num_copies', ascending=True)
df_sort.head()
```

### Use SQL to answer the following questions: 

1.  How many rows are in the 'author' database? 

```python
sql_str = """SELECT * FROM author
"""

df = pd.read_sql_query(sql_str, conn)
len(df)
```

2. How many authors are called David?  

```python
sql_str = """
SELECT * FROM author a
WHERE a.author LIKE 'David%'
"""

df_david = pd.read_sql_query(sql_str, conn)
len(df_david)
```

3. How many books each author wrote? 

```python
sql_str = """
SELECT author, count(book) as num_books 
FROM author
GROUP BY author
"""

df = pd.read_sql_query(sql_str, conn)
df.sort_values('num_books', ascending=False)
```

4. Use both the author table and books table to 
say: how many copies were sold by authors named David 


    a. how many books for each of the Davids
    
    
    b. how many books in total for all Davids

```python
sql_str_david = """
SELECT * FROM author a
WHERE a.author LIKE 'David%'
"""

sql_2 = """
SELECT author, sum(num_copies) as total_copies
FROM ({}) d
JOIN books b on d.book = b.books
GROUP BY author
""".format(sql_str_david)



df = pd.read_sql_query(sql_2, conn)
df.sort_values('total_copies', ascending=False)
```
