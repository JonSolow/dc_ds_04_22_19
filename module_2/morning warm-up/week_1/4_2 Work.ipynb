```python
import sqlite3
import pandas as pd
```

```python
cat_conn = sqlite3.connect('cats.db')
dog_conn = sqlite3.connect('dogs.db')
owners_conn = sqlite3.connect('owners.db')
```

```python
sql_str_1 = """

SELECT * FROM dogs

"""

df = pd.read_sql_query( sql_str_1, dog_conn)
df
```

```python
conn = sqlite3.connect('cats.db')
c = conn.cursor()
c.execute('ATTACH DATABASE "dogs.db" AS db_dogs')
c.execute('ATTACH DATABASE "owners.db" AS db_owners')

sql_str = """


SELECT U.Name
FROM
(SELECT *, "cat" as Species from cats c
UNION
SELECT *, "dog" as Species from db_dogs.dogs d) U
JOIN 
(SELECT * from db_owners.owners o
WHERE o.Name="Avi Flombaum") Avi
ON Avi.ID = U.Owner_ID


"""



df = pd.DataFrame(c.execute(sql_str).fetchall())
df.columns = [x[0] for x in c.description]
df
```
