```python
import numpy as np
from collections import Counter
```

```python
s1 = "a great game"
s2 = "the election was over"
s3 = "very clean match"
s4 = "a clean but forgettable game"
s5 = "it was a close election"
```

```python
s1 = s1.split(" ")
s2 = s2.split(" ")
s3 = s3.split(" ")
s4 = s4.split(" ")
s5 = s5.split(" ")
    
            
```

```python
sports = [s1, s3, s4]
```

```python
notsports = [s2, s5]
```

```python
flat_list = []
for sublist in sports:
    for item in sublist:
        flat_list.append(item)
        
sports = flat_list.copy()
```

```python
flat_list = []
for sublist in notsports:
    for item in sublist:
        flat_list.append(item)
        
notsports = flat_list.copy()
```

```python
sport_count = Counter(" ".join(sports))
notsport_count = Counter(" ".join(notsports))
```

```python
sport_laplace = {}
for k,v in sport_count.items():
    sport_laplace[k] = v + 1
    
notsport_laplace = {}
for k,v in notsport_count.items():
    notsport_laplace[k] = v + 1
```

```python
sport_laplace, notsport_laplace
```

```python
for k in sport_laplace.keys():
    if k not in notsport_laplace.keys():
        notsport_laplace[k] = 1

for k in notsport_laplace.keys():
    if k not in sport_laplace.keys():
        sport_laplace[k] = 1
```

```python
sports_probs = {}
for k,v in sport_laplace.items():
    sports_probs[k] = v / sum(sport_laplace.values())
    
notsport_probs = {}
for k,v in notsport_laplace.items():
    notsport_probs[k] = v / sum(notsport_laplace.values())
```

```python
sports_probs, notsport_probs
```

```python
p_sp_giv_avcg_num = sports_probs['a'] * sports_probs['very'] * sports_probs['close'] * sports_probs['game'] * (3/5)
p_sp_giv_avcg_num
```

```python
p_nsp_giv_avcg_num = notsport_probs['a'] * notsport_probs['very'] * notsport_probs['close'] * notsport_probs['game'] * (2/5)
p_nsp_giv_avcg_num
```

```python
p_sp_giv_avcg_num / (p_sp_giv_avcg_num + p_nsp_giv_avcg_num)
```
