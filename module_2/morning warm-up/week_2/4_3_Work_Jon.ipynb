We want to grab the name of the homeworld of the fourth pilot of the Millennium Falcon. How would we do so?

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://swapi.co/api')
r.status_code
```

```python
# population of genosis

planets_r = requests.get(r.json()['planets'])
```

```python
df_planets = pd.DataFrame.from_dict(planets_r.json()['results'])
df_planets.loc[df_planets['name']=='Geonosis', 'population']
```

```python
vehicles_r = requests.get(r.json()['vehicles'])
```

```python
df_vehicles = pd.DataFrame.from_dict(vehicles_r.json()['results'])
df_vehicles.loc[df_vehicles['name']=='AT-ST', "cargo_capacity"]
```

```python
def starwars_api(type_of_thing, name_of_thing=None, field_requested=None):
    r_temp = requests.get(r.json()[type_of_thing])
    df_temp = pd.DataFrame.from_dict(r_temp.json()['results'])
    if not name_of_thing:
        return df_temp
    elif not field_requested:
        return df_temp.loc[df_temp['name']==name_of_thing, :]
    else:
        return df_temp.loc[df_temp['name']==name_of_thing, field_requested]
```

```python
starwars_api('planets', 'Geonosis', 'population')
```

```python
starwars_api('vehicles', 'AT-ST', 'cargo_capacity')
```

```python
starwars_api('people', 'Darth Vader', 'birth_year')
```

```python
starwars_api('species', 'Gungan', 'eye_colors')
```

```python
url = list(starwars_api('starships', "Millennium Falcon", 'pilots').values)[0][3]
r_temp = requests.get(url).json()
url = r_temp['homeworld']
r_temp = requests.get(url).json()
r_temp['name']
```
